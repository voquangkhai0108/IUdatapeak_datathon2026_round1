```python
import pandas as pd
import os

# Đường dẫn
RAW_DIR = '../data/raw/'
PROCESSED_DIR = '../data/processed/'
```

```python
sales     = pd.read_csv(os.path.join(RAW_DIR, 'sales.csv'), parse_dates=['Date'])
sales = sales.dropna()

traffic   = pd.read_csv(os.path.join(RAW_DIR, 'web_traffic.csv'), parse_dates=['date'])
traffic = traffic.dropna()

shipment  = pd.read_csv(os.path.join(RAW_DIR, 'shipments.csv'), parse_dates=['ship_date', 'delivery_date'])
shipment = shipment.dropna()

review  = pd.read_csv(os.path.join(RAW_DIR, 'reviews.csv'), parse_dates=['review_date'])
review = review.dropna()

returns = pd.read_csv(os.path.join(RAW_DIR, 'returns.csv'), parse_dates=['return_date'])
returns = returns.dropna()

customer = pd.read_csv(os.path.join(RAW_DIR, 'customers.csv'), parse_dates=['signup_date'])
customer = customer.dropna()

inventory = pd.read_csv(os.path.join(RAW_DIR, 'inventory.csv'), parse_dates=['snapshot_date'])
inventory = inventory.dropna()

orders = pd.read_csv(os.path.join(RAW_DIR, 'orders.csv'), parse_dates=['order_date'])
orders= orders.dropna()
```

```python
# File mẫu để làm khung cho tập Test
test_df   = pd.read_csv(os.path.join(RAW_DIR, 'sample_submission.csv'), parse_dates=['Date'])
```

```python
# 2. Gom dữ liệu theo ngày (order_date)
# Giả sử bảng orders có cột 'order_id' và 'total_amount'
daily_orders = orders.groupby('order_date').agg({
    'order_id': 'nunique',       
    'customer_id': 'nunique'   
           
}).reset_index()

daily_orders = daily_orders.rename(columns={
    'order_id': 'number_of_order',
    'customer_id': 'number_of_customer'
})

daily_orders
```

```python
daily_sales = sales.groupby('Date').agg({
    'Revenue': 'sum',       
    'COGS': 'sum'    
           
}).reset_index()


daily_sales
```

```python
daily_traffic = traffic.groupby('date').agg({
    'sessions': 'sum',       
    'page_views': 'sum'
           
}).reset_index()


daily_traffic = daily_traffic.rename(columns={
    
    'sessions': 'sum_of_sessions',
    'page_views': 'sum_of_page_views'
})

daily_traffic
```

```python
daily_shipment = shipment.groupby('ship_date').agg({
    'shipping_fee': 'sum',       
        
}).reset_index()

daily_shipment = daily_shipment.rename(columns={

    'shipping_fee': 'sum_of_shipping_fee'
})

daily_shipment
```

```python
daily_review = review.groupby('review_date').agg({      
    'customer_id': 'nunique',    
    'rating'     : 'mean'    
}).reset_index()

daily_review = daily_review.rename(columns={
    
    'customer_id': 'number_visitor_review',
    'rating': 'mean_of_rating'
})

daily_review
```

```python
daily_returns = returns.groupby('return_date').agg({

    'refund_amount' : 'sum'   
           
}).reset_index()

daily_returns = daily_returns.rename(columns={
    'refund_amount': 'sum_of_refund',
   
})

daily_returns
```

```python
daily_customer = customer.groupby('signup_date').agg({
    'customer_id': 'nunique',        
           
}).reset_index()

daily_customer = daily_customer.rename(columns={
    
    'customer_id': 'number_of_customer_signup'
})

daily_customer
```

```python
daily_inventory = inventory.groupby('snapshot_date').agg({
    'stock_on_hand': 'sum',
    'units_sold': 'sum',   
    'units_received': 'sum'     
           
}).reset_index()

daily_inventory
```

```python
master_df = daily_sales.merge(daily_orders, left_on='Date',right_on= 'order_date',  how='left')
master_df = master_df.drop('order_date', axis=1)
master_df
```

```python
# Chuyển đổi toàn bộ cột Date sang datetime
master_df['Date'] = pd.to_datetime(master_df['Date'])

# Bây giờ mới thêm cột month
master_df['month'] = master_df['Date'].dt.month
```

```python
master_df = master_df.merge(daily_traffic, left_on='Date', right_on='date', how = 'left')
master_df = master_df.drop('date', axis=1)
master_df
```

```python
master_df = master_df.merge(daily_shipment, left_on='Date', right_on='ship_date', how = 'left')
master_df = master_df.drop('ship_date', axis=1)
master_df
```

```python
master_df = master_df.merge(daily_review, left_on='Date', right_on='review_date', how = 'left')
master_df
```

```python
master_df = master_df.drop('review_date', axis=1)
master_df
```

```python
master_df = master_df.merge(daily_returns, left_on='Date', right_on='return_date', how = 'left')
master_df
```

```python
master_df = master_df.merge(daily_customer, left_on='Date', right_on='signup_date', how = 'left')
master_df
```

```python
master_df = master_df.drop('return_date', axis=1)
master_df = master_df.drop('signup_date', axis=1)

master_df
```

```python
cols_to_fill = ['sum_of_sessions', 'sum_of_page_views', 'sum_of_shipping_fee', 'number_visitor_review', 'mean_of_rating', 'sum_of_refund', 'number_of_customer_signup']

# Fill tất cả trong 1 vòng lặp
for col in cols_to_fill:
    master_df[col] = master_df.groupby('month')[col].transform(
        lambda x: x.fillna(x.mean())
    )

master_df
```

```python
master_df.to_csv(os.path.join(PROCESSED_DIR, 'data_train.csv'), index=False)
print(" Hoàn thành: Đã làm sạch và gom tất cả về bảng data train !")
```
